```python
from pulp import *
```

```python
# Create LP Maximization problem
model = LpProblem("Maximize_Profit", LpMaximize)
```

```python
# Define variables
A = LpVariable('Product_A', lowBound=0)
B = LpVariable('Product_B', lowBound=0)
```

```python
# Maximize Profit
model += 20*A + 30*B
```

```python
# Constraints
model += 2*A + 4*B <= 100   # Labor
model += 3*A + 2*B <= 90    # Machine
model += A + B <= 40        # Material
```

```python
model.solve()
```

```python
print("Status:", LpStatus[model.status])

print("Optimal Production:")
print("Product A =", A.varValue)
print("Product B =", B.varValue)

print("Maximum Profit = ₹", value(model.objective))
```
